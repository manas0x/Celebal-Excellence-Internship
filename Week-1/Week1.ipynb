```python
# Check missing values
ds.isnull().sum()
```

```python
#Handle Missing Values
ds["discount"] = ds["discount"].fillna(0)

ds["what_customers_said"] = ds["what_customers_said"].fillna("No Review")

ds["seller_name"] = ds["seller_name"].fillna("Unknown Seller")

ds["videos"] = ds["videos"].fillna("No Video")

ds["seller_information"] = ds["seller_information"].fillna("Not Available")

ds["variations"] = ds["variations"].fillna("No Variations")
```

```python
ds.isnull().sum()
```

```python
#Duplicate Record Check
print("Duplicate Record : " , ds.duplicated().sum())
```

```python
print(ds["initial_price"].dtype)
print(ds["final_price"].dtype)
```

```python
# Show first 5 rows of initial_price and final_price
ds[["initial_price", "final_price"]].head()
```

```python
ds["final_price"] = pd.to_numeric(
    ds["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False),
    errors="coerce"
)
```

```python
ds["discount_amount"] = ds["initial_price"] - ds["final_price"]
```

```python
ds[["initial_price", "final_price", "discount_amount"]].head()
```
